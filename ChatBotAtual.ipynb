```python
from chatterbot import ChatBot
from chatterbot.trainers import ListTrainer


from spacy.cli import download

download("en_core_web_sm")

class ENGSM:
    ISO_639_1 = 'en_core_web_sm'
```

```python

chatbot = ChatBot("BotEasy", tagger_language=ENGSM)

conversa = [
    "Boa tarde",
    "Boa tarde, qual o seu pedido?",
    "Boa noite",
    "Boa noite, qual o seu pedido?",
    "Olá",
    "Olá!, qual o seu pedido?",
    "p01",
    "Pizza de mussarela escolhida!, algo mais?",
    "s",
    "Digite mais um item:",
    "sim",
    "Digite mais um item:",
    "p02",
    "Pizza de calabresa escolhida!, algo mais?",
    "s01",
    "Suco de laranja escolhido!, algo mais?",
    "s02",
    "Suco de uva escolhido!, algo mais?",
    "Pizza de mussarela",
    "Pizza de mussarela escolhida!, algo mais?",
    "Pizza de calabresa",
    "Pizza de calabresa escolhida!, algo mais?",
    "Suco de laranja",
    "Suco de laranja escolhido!, algo mais?",
    "Suco de uva",
    "Suco de uva escolhido!, algo mais?"
    
]

trainer = ListTrainer(chatbot)
trainer.train(conversa)
```
